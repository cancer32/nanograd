```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
    return (x**2)-(5*x)+3
```

```python
xs = np.arange(-5, 5, step=0.25)
ys = f(xs)
plt.plot(xs, ys)
plt.show()
```

```python
h = 0.000000001
x = 2.5
(f(x+h)-f(x))/h
```
